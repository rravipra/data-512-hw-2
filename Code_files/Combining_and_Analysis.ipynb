# Combining the Datasets:

```python
# import the required libraries
import pandas as pd
from tabulate import tabulate
```

Extracting the data as dataframes from the CSV files:

```python
# get all the data that we have as CSV files (created by the previous code and also the ones that are provided)
df_ores_scores = pd.read_csv('/content/drive/MyDrive/ores_scores.csv') # created in get_ores_scores.ipynb
df_city_by_state = pd.read_csv('/content/drive/MyDrive/us_cities_by_state_SEPT.2023.csv') # provided and the link can be found in the ReadMe file
df_population = pd.read_excel('/content/drive/MyDrive/NST-EST2022-POP.xlsx', header=[2,3]) # link provided in the ReadMe file
df_states_region = pd.read_excel('/content/drive/MyDrive/US States by Region - US Census Bureau.xlsx') # provided and the link can be found in the ReadMe file
```

The dataframe containing the ORES scores.

```python
# this is the dataframe of the ores scores that was extracted as a csv file through the LiftWing ML Service API
df_ores_scores
```

The dataframe of the city by state:

```python
# the dataframe of the city by state:
df_city_by_state
```

Population dataframe by Geographic area and the states:

```python
# the dataframe of the geographic area and the population estimates of the years 2020, 2021 and 2022 where we only need the one from 2022 when we merge with the other datasets.
df_population
```

```python
# Remove any leading dots in the 'Geographic Area' column to match the 'state' column
df_population.columns = ['Geographic Area', 'April 1, 2020', '2020', '2021', '2022']

# We just need the latest population that is from 2022
population_df = df_population[['Geographic Area', '2022']].rename(columns={'2022': 'Population'})

# Doing data manipulation to just extract the required rows (Please check out the CSV file which will give you an idea of why I am using this particular code to extract the Geographic area and the population)
population_df = population_df.iloc[:58,:].dropna().reset_index().drop(columns = 'index')

population_df
```

The dataframe of the states by region:

```python
# use fillna to fill the NA values with the Division and the Region using the 'ffill' method
# We use this since in the CSV file the every region and division are only shown once (i.e for a set of states if it belongs to a particular division and a region
# for all those states the region and division are mentioned only once)

df_states_region["REGION"].fillna(method="ffill", inplace=True)
df_states_region["DIVISION"].fillna(method="ffill", inplace=True)

# reset the index
df_states_region = df_states_region.dropna().reset_index().drop(columns = 'index')
df_states_region
```

```python
# Define a mapping for state name conversions (this is to make sure that we are able to merge with the df_ores_scores dataframe
# and for the states to have the same names without any additional punctuations)

state_name_mapping = {
    'Georgia_(U.S._state)': 'Georgia',
    'New_Hampshire': 'New Hampshire',
    'New_Jersey': 'New Jersey',
    'New_Mexico': 'New Mexico',
    'New_York': 'New York',
    'North_Carolina': 'North Carolina',
    'North_Dakota': 'North Dakota',
    'Rhode_Island': 'Rhode Island',
    'South_Carolina': 'South Carolina',
    'South_Dakota': 'South Dakota',
    'West_Virginia': 'West Virginia'
}

# Apply the state name mapping to map the state names to the appropriate ones mentioned above:

df_city_by_state['state'] = df_city_by_state['state'].replace(state_name_mapping)
```

Merge all the dataframes together to get the desired dataframe

```python
# Clean state names in population_df using the strip function (please refer to population_df above to see why I am doing this)
population_df['Geographic Area'] = population_df['Geographic Area'].str.strip('.').str.strip()

# Clean state names in df_city_by_state
df_city_by_state['state'] = df_city_by_state['state'].str.strip()

df_ores_scores['rev_id'] = df_ores_scores['rev_id'].astype(str)

# Merge df_city_by_state with df_ores_scores
merged_df = df_city_by_state.merge(df_ores_scores, left_on="page_title", right_on="title", how="left")

# Merge with population_df to add population data
merged_df = merged_df.merge(population_df, left_on="state", right_on="Geographic Area", how="left")

# Merge with df_states_region to add regional-division data
merged_df = merged_df.merge(df_states_region, left_on="state", right_on="STATE", how="left")

# Identify unmatched areas
unmatched_areas = merged_df[merged_df["DIVISION"].isna()]["state"].unique()

# Filter out non-state entries
valid_states = merged_df[~merged_df["state"].isin(unmatched_areas)]

# Create the final consolidated DataFrame
final_df = valid_states[["state", "DIVISION", "Population", "page_title", "rev_id", "prediction"]]
```

```python
final_df = final_df.drop_duplicates().reset_index().drop(columns = 'index').dropna()
final_df = final_df.rename(columns = {'DIVISION':'regional_division', 'Population': 'population', 'page_title': 'article_title', 'rev_id':'revision_id',
                           'prediction': 'article_quality'})
final_df = final_df[['state', 'regional_division', 'population', 'article_title', 'revision_id', 'article_quality']]
```

```python
final_df.to_csv('wp_scored_city_articles_by_state.csv', index=False)
```

#Analysis:

The analysis will consist of calculating total-articles-per-population (a ratio representing the number of articles per person)  and high-quality-articles-per-population (a ratio representing the number of high quality articles per person) on a state-by-state and divisional basis. All of these values are “per capita” ratios.
Consider "high quality" articles to be articles that ORES predicted would be in either the "FA" (featured article) or "GA" (good article) classes.


```python
# Define what high quality articles are in this case: We are considering "high quality" articles to be articles that ORES predicted
# would be in either the "FA" (featured article) or "GA" (good article) classes
high_quality_articles = final_df[final_df['article_quality'].isin(['FA', 'GA'])]

# Group by state and count articles
state_article_counts = final_df.groupby('state').size()
state_high_quality_article_counts = high_quality_articles.groupby('state').size()
```

```python
state_population = final_df.drop_duplicates(subset='state')[['state', 'population']].set_index('state')

# calculate the per capita ratios for each state based on the number of articles
state_articles_per_capita = state_article_counts / state_population['population']

# calculate the per capita ratios for each stata based on the high quality articles
state_high_quality_articles_per_capita = state_high_quality_article_counts / state_population['population']
```

#Questions to answer: Produce tables for

1) Top 10 US states by coverage: The 10 US states with the highest total articles per capita (in descending order).

2) Bottom 10 US states by coverage: The 10 US states with the lowest total articles per capita (in ascending order).

3) Top 10 US states by high quality: The 10 US states with the highest high quality articles per capita (in descending order).

4) Bottom 10 US states by high quality: The 10 US states with the lowest high quality articles per capita (in ascending order).

5) Census divisions by total coverage: A rank ordered list of US census divisions (in descending order) by total articles per capita.

6) Census divisions by high quality coverage: Rank ordered list of US census divisions (in descending order) by high quality articles per capita.#

```python
# Get the top 10 states by coverage
top_10_states_by_coverage = state_articles_per_capita.sort_values(ascending=False).head(10)

# get the bottom 10 states by coverage
bottom_10_states_by_coverage = state_articles_per_capita.sort_values(ascending=True).head(10)

# get the top 10 states by high quality
top_10_states_by_high_quality = state_high_quality_articles_per_capita.sort_values(ascending=False).head(10)

# get the bottom 10 states by high quality
bottom_10_states_by_high_quality = state_high_quality_articles_per_capita.sort_values(ascending=True).head(10)
```

#1) Top 10 states by coverage

```python
# Convert the series into a DataFrame and reset the index
df_top_10 = top_10_states_by_coverage.reset_index()
df_top_10.columns = ['State', 'Coverage']

# Make the rank as the index
df_top_10.index = range(1, len(df_top_10) + 1)

# Print using tabulate
print(tabulate(df_top_10, headers='keys', tablefmt='pipe'))
```

#2) Bottom 10 States by coverage:

```python
# Convert the series into a DataFrame and reset the index
df_bottom_10 = bottom_10_states_by_coverage.reset_index()
df_bottom_10.columns = ['State', 'Coverage']

# Make the rank as the index
df_bottom_10.index = range(1, len(df_bottom_10) + 1)

# Print using tabulate
print(tabulate(df_bottom_10, headers='keys', tablefmt='pipe'))
```

# 3) Top 10 states by high quality

```python
# Convert the series into a DataFrame and reset the index
df_top_10_high_quality = top_10_states_by_high_quality.reset_index()
df_top_10_high_quality.columns = ['State', 'Quality']

# Make the rank as the index
df_top_10_high_quality.index = range(1, len(df_top_10_high_quality) + 1)

# Print using tabulate
print(tabulate(df_top_10_high_quality, headers='keys', tablefmt='pipe'))
```

# 4) Bottom 10 states by high quality

```python
# Convert the series into a DataFrame and reset the index
df_bottom_10_high_quality = bottom_10_states_by_high_quality.reset_index()
df_bottom_10_high_quality.columns = ['State', 'Quality']

# Make the rank as the index
df_bottom_10_high_quality.index = range(1, len(df_bottom_10_high_quality) + 1)

# Print using tabulate
print(tabulate(df_bottom_10_high_quality, headers='keys', tablefmt='pipe'))
```

We need to do some data manipulation which is to get the sum of the populations based on the divsion and then calculating the per capita ratios which is done below:

```python
# Get the unique population for each state
state_population = final_df.drop_duplicates(subset='state').set_index('state')['population']

# Compute the total population for each division
division_population = final_df.drop_duplicates(subset='state').groupby('regional_division')['population'].sum()

# Group by regional_division and count total and high-quality articles
division_article_counts = final_df.groupby('regional_division').size()
division_high_quality_article_counts = high_quality_articles.groupby('regional_division').size()

# The articles-per-capita and high-quality-articles-per-capita for each division
division_articles_per_capita = division_article_counts / division_population
division_high_quality_articles_per_capita = division_high_quality_article_counts / division_population

# Rank divisions by articles per capita and high-quality articles per capita
division_rank_by_total = division_articles_per_capita.sort_values(ascending=False)
division_rank_by_hq = division_high_quality_articles_per_capita.sort_values(ascending=False)
```

#5) Census division by total coverage

```python
# Convert the series to a list of tuples with rank
table_data_total = [(index+1, division, value) for index, (division, value) in enumerate(division_rank_by_total.items())]

# Print the table
print("Divisions ranked by total articles per capita:\n")
print(tabulate(table_data_total, headers=['Rank', 'Division', 'Articles per Capita'], tablefmt='pipe', floatfmt=("", ".", ".6f")))
```

#6) Census division by high quality coverage

```python
# Convert the series to a list of tuples with rank
table_data_total_by_hq = [(index+1, division, value) for index, (division, value) in enumerate(division_rank_by_hq.items())]

# Print the table
print("Divisions ranked by high quality articles per capita:\n")
print(tabulate(table_data_total_by_hq, headers=['Rank', 'Division', 'Articles per Capita'], tablefmt='pipe', floatfmt=("", ".", ".6f")))
```
